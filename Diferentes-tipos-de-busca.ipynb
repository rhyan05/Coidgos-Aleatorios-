```python
import numpy as np

from tqdm import tqdm
from datetime import datetime

import matplotlib.pyplot as plt
%matplotlib inline
```

# Implementações dos algoritmos de ordenação

```python
def insertion_sort(lista):
  n = len(lista)

  for i in tqdm(list(range(1, n)), ncols=50):
    elemento = lista[i]
    j = i - 1

    while (j >= 0) and (elemento < lista[j]):
      lista[j + 1] = lista[j]
      j -= 1
    lista[j + 1] = elemento

np.random.seed(0)
teste = list(np.random.rand(10000))
insertion_sort(teste)
#teste
```

```python
def selection_sort(lista):
  n = len(lista)

  for i in tqdm(list(range(n - 1)), ncols=50):
    menor = i

    for j in range(i + 1, n):
      if lista[j] < lista[menor]:
        menor = j

    if i != menor:
      aux = lista[i]
      lista[i] = lista[menor]
      lista[menor] = aux

np.random.seed(0)
teste = list(np.random.rand(10000))
selection_sort(teste)
#teste
```

```python
def bubble_sort(lista):
  n = len(lista)

  for i in tqdm(list(range(n - 1, 0, -1)), ncols=50):
    for j in range(1, i + 1):
      if lista[j - 1] > lista[j]:
        aux = lista[j - 1]
        lista[j - 1] = lista[j]
        lista[j] = aux

np.random.seed(0)
teste = list(np.random.rand(10000))
bubble_sort(teste)
#teste
```

# Testes dos algoritmos de ordenação

1. Conferir se a ordenação está correta
2. Exibir um gráfico _quantidade de elementos_ $\times$ _tempo de processamento_ para cada algoritmo

```python
N_max_elementos = 10000
N_min_elementos = 1000
N_steps = 500

quantidades_de_elementos = list(range(N_min_elementos,
                                      N_max_elementos + 1,
                                      N_steps))

print('Testando algoritmo Insertion sort')
np.random.seed(0)
tempo_de_processamento_ins = []
for n_elementos in quantidades_de_elementos:
  lista = list(np.random.rand(n_elementos))

  inicio = datetime.now()
  insertion_sort(lista)
  tempo = datetime.now() - inicio
  tempo_de_processamento_ins.append(tempo.total_seconds())

print()
print('Testando algoritmo Selection sort')
np.random.seed(0)
tempo_de_processamento_sel = []
for n_elementos in quantidades_de_elementos:
  lista = list(np.random.rand(n_elementos))

  inicio = datetime.now()
  selection_sort(lista)
  tempo = datetime.now() - inicio
  tempo_de_processamento_sel.append(tempo.total_seconds())

print()
print('Testando algoritmo Bubble sort')
np.random.seed(0)
tempo_de_processamento_bub = []
for n_elementos in quantidades_de_elementos:
  lista = list(np.random.rand(n_elementos))

  inicio = datetime.now()
  bubble_sort(lista)
  tempo = datetime.now() - inicio
  tempo_de_processamento_bub.append(tempo.total_seconds())
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.title('Desempenho do Insertion sort')
plt.xlabel('Quantidades de elementos')
plt.ylabel('Tempo de processamento')
plt.plot(quantidades_de_elementos, tempo_de_processamento_ins)
plt.grid()

plt.subplot(1,3,2)
plt.title('Desempenho do Selection sort')
plt.xlabel('Quantidades de elementos')
plt.ylabel('Tempo de processamento')
plt.plot(quantidades_de_elementos, tempo_de_processamento_sel)
plt.grid()

plt.subplot(1,3,3)
plt.title('Desempenho do Bubble sort')
plt.xlabel('Quantidades de elementos')
plt.ylabel('Tempo de processamento')
plt.plot(quantidades_de_elementos, tempo_de_processamento_bub)
plt.grid()

plt.tight_layout();
```
